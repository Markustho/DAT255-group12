Categorical columns are handled very differently in neural networks, compared to decision tree approaches. As we saw in <<chapter_collab>>, in a neural net a great way to handle categorical variables is by using embeddings. To create embeddings, fastai needs to determine which columns should be treated as categorical variables. It does this by comparing the number of distinct levels in the variable to the value of the `max_card` parameter. If it's lower, fastai will treat the variable as categorical. Embedding sizes larger than 10,000 should generally only be used after you've tested whether there are better ways to group the variable, so we'll use 9,000 as our `max_card`:

```python
cont_nn,cat_nn = cont_cat_split(df_nn_final, max_card=9000, dep_var=dep_var)
```

# Using ensemble

## Check test set RMSE using ensemble

```python
def ensemble():
    learn = tabular_learner(dls,y_range=(19,120),layers=[500,250],
                        n_out=1,loss_func=F.mse_loss)
    with learn.no_bar(),learn.no_logging(): learn.fit_one_cycle(5,1e-2)
    preds, targets = learn.get_preds()
    return preds, targets
```

```python
# Create a list of predictions and targets from multiple learners
results = [ensemble() for _ in range(5)]

# Separate the predictions and the actual targets
learns = [res[0] for res in results]  # List of predictions
targs_list = [res[1] for res in results]  # List of target batches

# Assuming that the targets are the same for each model's predictions
# (which they should be if the DataLoader is the same)
ens_targs = targs_list[0]  # Take the targets from the first model

# Stack and average the predictions
ens_preds = torch.stack(learns).mean(0)
```

```python

print(f'root mean square error: {root_mean_squared_error(ens_preds,ens_targs)}')
```

Vi fikk litt bedre RMSE ved å bruke ensemble
